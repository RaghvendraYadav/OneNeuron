```python
import os
import csv
import logging as lg
import mysql.connector as connection
```

```python
#remove the log filr if it exist
if os.path.exists('database.log'):
    os.remove('database.log')
```

```python
lg.basicConfig(filename='database.log',level=lg.INFO,format='%(asctime)s - %(message)s')
```

```python
console_log=lg.StreamHandler()
console_log.setLevel(lg.DEBUG)
format=lg.Formatter('%(asctime)s - %(message)s')
console_log.setFormatter(format)
lg.getLogger('').addHandler(console_log)
lg.info('Logging Started')
try:
    con=connection.connect(host='localhost',user='root',password='0000',database='student1')
    if con.is_connected():
        lg.info('Establishing the connection ... ....')
        lg.info('Connection Established in the database!')
        cursor=con.cursor()
        cursor.execute('Drop table If exists carbon_nanotubes')
        lg.info('creating table-----------')
        cursor.execute("Create table carbon_nanotubes(Chiral_Indice_n int(7),"
                      "Chiral_Indice_m int(7),"
                       "Initial_atomic_coordinates_u int(7),"
                       "Initial_atomic_coordinates_v int(7),"
                       "Initial_atomic_coordinates_w int(7),"
                       "Calculateda_tomic_coordinates_u int(7),"
                       "Calculateda_tomic_coordinates_v int(7),"
                       "Calculateda_tomic_coordinates_w int(7));"
                )
        lg.info('Table Created usccessfully')
        
        
        with open('carbon_nanotubes.csv','r') as file:
            next(file)
            carbon_data=csv.reader(file,delimiter='\n')
            for num,record in enumerate(carbon_data,start=1):
                rec=record[0].split(';')
                record=[]
                for i in rec:
                    if len(i.split(','))>1:
                        record.append(int(i.split(',')[1]))
                    else:
                        record.append(int(i))
                record=','.join([str(elem) for elem in record])  
                cursor.execute(
                    "insert into carbon_nanotubes values({values})".format(values=record)
                )
                lg.info(f'Record{num} inerted')
            lg.info('All records inserted')
            con.commit()
            
except Exception as e:
    lg.error(str(e))
    con.close()
    
```
