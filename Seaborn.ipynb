```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)# Relational Plot  in Seaborn 
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex')# Group the data using Categorical Column 
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker')#  hue grouping the data and styple will give different Symbols
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker',size='size')# Size of dot will increase
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',size='size',col='day')
```
