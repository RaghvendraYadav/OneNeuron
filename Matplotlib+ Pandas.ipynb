```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(20)
df=pd.DataFrame(np.random.randn(10,4),index=pd.date_range('2021-07-18',periods=10),columns=['a','b','c','d'])
```

```python
df
```

```python
# New Column
df['New']=df['c']+df['d']
```

```python
df
```

```python
df['new_col_sqr_d']=df['d'].apply(lambda x:x*x)
```

```python
df
```

```python
df['new_col_sqr_d_log']=df['d'].apply(lambda x:np.log(x))
```

```python
df
```

```python
def test(x):
    if x<0:
        return x
    else:
        return np.log(x)
```

```python
df['new_col_sqr_log1']=df['d'].apply(test)# apply is used to use user defiend and inbuilt function for particular column 
```

```python
df
```

```python
def Test1(x):
    if x<0:
        return 1
    else:
        return 0
```

```python
df['new_col_01']=df['a'].apply(Test1)
```
